```python
#Convertir en dataframe
df= pd.read_csv("CountryTable.csv")
#df
```

```python
#Imprimimos las primeras filas o registros del Dataframe
df.head(5)
```

**FILTROS BASICOS PARA VARIABLES CUALITATIVAS **

```python
#Filtro por objeto

#Aqui se arrojan solo los datos de pa√≠ses que esten en Europa
filtro1=df[df["continent"] == "Europe"]
filtro1
```

```python
filtro1.to_csv("Filtro1.csv")
```

```python
#Filtro por filas

#Aqui se arrojan datos de la fila 10 a la 16
filtro2=df.iloc[ 10:16 , : ]  #filas no consecutivas
filtro2
```

```python
filtro2.to_csv("Filtro2.csv")
```

```python
#Filtro por columnas

#Aqui se arrojan datos de la fila 1, 4 y 6
filtro3=df.iloc[ : , [1,4,6]]  #columnas no consecutivas
filtro3
```

```python
filtro3.to_csv("Filtro3.csv")
```

```python
df= pd.read_csv("CountryTable.csv", index_col=1)
df
#Aqui establecemos la columna "name" como el index
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas

#Aqui obtenemos datos de la poblacion de mexico y espa√±a
filtro4=df.loc[["Mexico","Spain"], ["population"]]
filtro4
```

```python
filtro4.to_csv("Filtro4.csv")
```

```python
#Filtro por cabecera

#Aqui solo nos arroja las primeras 10 filas
filtro5=df.head(10)
filtro5
```

```python
filtro5.to_csv("Filtro5.csv")
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion

#Aqui solo observamos datos de pa√≠ses con poblacion mayor a 100 millones
filtro6=df[df["population"] > 100000000]
filtro6
```

```python
filtro6.to_csv("Filtro6.csv")
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y

#Aqui solo observamos datos de pa√≠ses con poblacion mayor a 100 millones y que esten en √Åsia

filtro7=df[(df["population"] > 100000000) & (df["continent"] == "Asia")]
filtro7
```

```python
filtro7.to_csv("Filtro7.csv")
```

```python
#Filtro o

#Aqui solo observamos datos de pa√≠ses con poblacion mayor a 100 millones o que esten en √Åsia

filtro8=df[(df["population"] > 100000000) | (df["continent"] == "Asia")]
filtro8
```

```python
filtro8.to_csv("Filtro8.csv")
```

```python
#Filtro not

#Aqui no observamos datos de pa√≠ses con poblacion mayor a 100 millones y que esten en √Åsia

filtro9=df[~(df["population"] >= 100000000) & ~(df["continent"] == "Asia")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Filtro9.csv")
```
